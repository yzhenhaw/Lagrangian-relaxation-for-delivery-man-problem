```python
person=[]
supply=[]
demand=[]
for i in range(1,51):
    person.append(i)
for j in range(51,101):
    supply.append(j)
for k in range(101,151):
    demand.append(k)
print(person,supply,demand)
```

```python
#person param
print("param: P: person :=")
p_data=[]
for p in person:
    pn=1
    p_data.append(pn)

for i in range(1,51):
    pair='p'+str(i)
    print(pair,p_data[i-1])
print(";")
# supply param

def generate_node(supply,demand):
    xp=[]
    xs=[]
    xd=[]
    sums=0
    sumd=0
    for i in supply:
        s=random.randint(1,5)
        xs.append(s)
        sums+=s
    for k in demand:
        dn=random.randint(1,5)
        xd.append(dn)
        sumd+=dn       
    return xs,xd,sums,sumd

s_data,d_data,s_sum,d_sum=generate_node(supply,demand)
while s_sum!=d_sum:
    s_data,d_data,s_sum,d_sum=generate_node(supply,demand)
print("param: O: supply :=")    
for j in range(51,101):
    pair='o'+str(j)
    print(pair,s_data[j-51])
print(";")
print("param: D: demand :=")
for k in range(101,151):
    pair='d'+str(k)
    print(pair,d_data[k-101])
print(";")
```
